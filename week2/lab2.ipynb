- Use regular expression and lambda function to parse data.

```python
import pandas as pd
import re
```

```python
df = pd.read_csv('sales_fixed.csv')
```

```python
pattern = re.compile(r'"(.+)"\s(.+)')
```

```python
df = df.apply(lambda row: pd.Series({
    "order_id": row["order_id"],
    "product_type": pattern.match(str(row["name"])).group(1) if pattern.match(str(row["name"])) else None,
    "product_name": pattern.match(str(row["name"])).group(2) if pattern.match(str(row["name"])) else None,
    "ordered_at": pd.to_datetime(str(row["ordered_at"])),
    "price": float(str(row["price"]).replace("$", "")),
    "quantity": int(row["quantity"]),
    "line_total": float(str(row["line_total"]).replace("$", ""))
}), axis=1)
```

```python
df
```
