5. Object Oriented Method

```python
#create figure (empty canvas)
fig = plt.figure()
#add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])#left, bottom, width, height
#plot on thatset of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set x Label')#notice the use of set_to begin mothods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
#create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])#main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])#inset axes

#large figure axes1
axes1.plot(x, y, 'b')
axes1.set_xlabel('x_label_axes2')
axes1.set_ylabel('y_label_axes2')
axes1.set_title('Axes 2 Title')

#large figure axes2
axes2.plot(x, y, 'r')
axes2.set_xlabel('x_label_axes2')
axes2.set_ylabel('y_label_axes2')
axes2.set_title('Axes 2 Title')
```

6. Suplot()

use similar to plt.figure() except use tuple unpacking to grab fig and axes

```python
fig, axes  = plt.subplots()
#now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
#empty canvasof 1 by 2 subplots
fig, axes  = plt.subplots(nrows= 1, ncols= 2)
```

```python
axes
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
#display the figure object
fig 
```

11. Legends

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

12. Plot range

```python
fig, axes = plt.subplots(1, 3, figsize= (12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes ranges");
```

Continuing with the job market data, use Matplotlib/Seaborn to visualize:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('job-market.csv')
```

```python
jobs_by_location = data['Classification'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=jobs_by_location.values, y=jobs_by_location.index)
plt.title('Number of Job Classification by Location')
plt.xlabel('Number of Jobs')
plt.ylabel('Classification')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
